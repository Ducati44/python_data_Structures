```python
import pandas as pd
import numpy as np
import seaborn as sns
```

```python
iris = sns.load_dataset('iris')
iris.head()
```

```python
del iris['species']
iris.head()
```

```python
iris['sepal_sum'] = iris['sepal_length'] + iris['sepal_width']
iris.head()
```

```python
iris['petal_sum'] = iris['petal_length'] + iris['petal_width']
iris.head()
```

```python
summary_dict = pd.DataFrame({'mean': iris.mean(),
                            'std': iris.std(),
                            'min': iris.min(),
                            'max': iris.max()})
summary_dict
```
